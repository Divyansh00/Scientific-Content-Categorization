```python
from nltk.tokenize import word_tokenize
```

```python
text = []
for category in set(df_work["discipline"].values):
    for i in range(df_work[df_work["discipline"]==category].shape[0]):
        print("Processing ",i," of 603")
        text.append((word_tokenize(df_work.iloc[i,1]), category))
```

```python
text[0]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
stop = stopwords.words('english')
stop += list(string.punctuation)
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_text(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
clean_texts = [(clean_text(t), category) for t, category in text]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
categories = [category for text, category in clean_texts]
```

```python
text_cat = [" ".join(text) for text, category in clean_texts]
```

```python
count_vec = CountVectorizer(max_features = 5000, ngram_range=(1,3), max_df=0.8)
x_train_features = count_vec.fit_transform(text_cat)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
print(x_train_features.todense())
```

```python
from sklearn.externals import joblib
joblib.dump(count_vec, 'weight_model.pkl')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_train_features,categories,test_size=0.1,random_state=0)
```

```python
parameters = {'gamma':[0.0001,0.001,0.00001], 'C':[1,0.1,10]}
svc = SVC()
clf = GridSearchCV(svc, parameters, verbose=10, n_jobs=-1)
clf.fit(x_train_features,categories)
```

```python
from sklearn.metrics import accuracy_score
```

```python
RF = RandomForestClassifier(n_estimators=500,n_jobs=-1)
RF.fit(x_train_features, categories)
```

```python
Y_predict_RF = RF.predict(X_test)
print("Accuracy score for Random Forest :", accuracy_score(Y_test, Y_predict_RF))
```
